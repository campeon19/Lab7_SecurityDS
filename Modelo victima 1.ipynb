```python
# Dependencias
#https://adversarial-robustness-toolbox.org/

import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten
import numpy as np
from art.utils import load_dataset
```

```python
#Se utilizara una version del dataset del MNIST con 60k digitos a mano que incluye ART
(train_images, train_labels), (test_images, test_labels), min, max = load_dataset(name="mnist")
```

```python
#Se utilizaran las primeras 50k observaciones para entrenar el modelo
train_images_original = train_images[:50000]
train_labels_original = train_labels[:50000]
```

```python
#Modelo para la clasificaciÃ³n de digitos
def create_model():
    # Defining the model
    model = tf.keras.models.Sequential([
        Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=(28, 28, 1)),
        MaxPool2D(pool_size=2),
        Conv2D(filters=64, kernel_size=3, activation="relu"),
        MaxPool2D(pool_size=2),
        Flatten(),        
        Dense(units=10, activation="softmax")
    ])

    # Compiling the model
    model.compile(
        optimizer="adam",
        loss="categorical_crossentropy",
        metrics=["accuracy"]
        )

    # Returning the model
    return model
```

```python
#Entrenamiento del modelo
model = create_model()

# Training the model
model.fit(
    x=train_images_original, 
    y=train_labels_original, 
    epochs=20,
    batch_size=256)
```

```python
#Guardamos el modelo
model.save("modelo_victima1")
```
